### loc_og_t2t_mou_6 (col-28) to loc_og_t2c_mou_9(col-43)
The above columns are summed up in the column loc_og_mou_6 to loc_og_mou_9. So these can be deleted

```python
### deleting the unnecessary columns which are aggregating to the monthly minutes column
telecom = telecom.drop(telecom.iloc[:,28:44].columns, axis=1)
telecom = telecom.drop(telecom.iloc[:,32:48].columns, axis=1)
telecom = telecom.drop(telecom.iloc[:,52:64].columns, axis=1)
telecom = telecom.drop(telecom.iloc[:,56:72].columns, axis=1)
telecom = telecom.drop(telecom.iloc[:,4:8].columns, axis=1)
```

### Replacing missing values with 0 for the onnet_mou and offnet_mou columns

```python
values_impute_0 = {}

#get the numeric columns from the telecom dataset
num_cols = telecom.select_dtypes(include=np.number).columns.to_list()

for val in num_cols : 
    values_impute_0[val] = 0

values_impute_0
```

```python
#impute the values with the default / expected values
telecom = telecom.fillna(value=values_impute_0)
```
